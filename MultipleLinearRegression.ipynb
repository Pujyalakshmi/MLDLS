# Import Libraries

```python
#import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Dataset

```python
#import Dataset
dataset = pd.read_csv('Startups.csv')
print(dataset)
```

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values

print(X)
print(Y)
```

```python
len(dataset)
```

```python
plt.scatter(dataset['Marketing Spend'], dataset['Profit'])
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.title('Marketing Spend vs. Profit')
plt.show()
```

```python
#Create dummy variables for the categorical variable State
dataset['NewYork_State'] = np.where(dataset['State']=='New York',1,0)
dataset['California_State'] = np.where(dataset['State']=='California',1,0)
dataset['Florida_State'] = np.where(dataset['State']=='Florida',1,0)

#Drop the original State from the dataframe
dataset.drop(columns=['State'],axis=1,inplace=True)
```

```python
dataset.head()
```

# Split the data

```python
#Split the data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3)
print(X_train.shape, X_test.shape)
```

# Fitting Simple Linear Regression

```python
#Fitting Simple Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

# Predicting the test results

```python
#Predicting the test results
Y_pred = model.predict(X_test)
print(Y_pred)
print(Y_train)
```

# Calculate Accuracy

```python
#Calculate Accuracy
accuracy = model.score(X_test,Y_test)
print(accuracy)
```

# Plot the Regressor Curve


```python
plt.scatter(dataset['Marketing Spend'], dataset['Profit'])
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.title('Marketing Spend vs. Profit (Training Set)')
plt.plot(X_train, model.predict(X_train), color='green')
plt.show()
```

```python
plt.scatter(dataset['Marketing Spend'], dataset['Profit'])
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.title('Marketing Spend vs. Profit (Test Set)')
plt.plot(X_test, model.predict(X_test), color='red')
plt.show()
```
